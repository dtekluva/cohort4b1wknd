## PLEASE READ HERE.
Looking at the cells in general overview reveals that we have 2380 values in total, with only 2 supposedly numeric columns, but some few error exists as follows:

 1. In the style column there are two omissions
 2. in the stars column there are some columns with the text "unrated" instead of values, this makes our pandas reader display these columns as non numeric columns.

<u>**First - Solutions:**</u>

 - Replace all "unrated" cells to either zero or to the mean of the
   overall column
   
 - convert the column into a floating point column and then retry describing
 - Since the style column isnumeric an rather impossible to extrapolate at first glance, simply remove all rows with null "style" cells.


```python
corrected_Stars_column = data.Stars.replace("Unrated", 0)
```

```python
data["corrected_stars"] = corrected_Stars_column.astype("float")
```
