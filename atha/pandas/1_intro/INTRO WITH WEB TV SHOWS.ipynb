```python
# CREATE A NEW EXCEL FILE WITH SHEETS REPRESENTING EACH OF THE YEARS OF MOVIES RELEASE 
# AND THE DATA AS ALL MOVIES RELEASED IN SAID YEAR

yearly_movies_grouped = shows_data.groupby(by = "Year Released")
```

```python
yearly_movies_grouped.groups # SHOWS ALL GROUPS RAW WITH JUST INDEX
# NOTE THAT WHAT IS RETURNED BY THE  (.GROUPS) METHOD IS ACTUALLY A DICTIONARY 
# WITH KEY OF THE GROUP NAME AND VALUE OF THE GROUP DATA AS A DATAFRAME
```

```python
yearly_movies_grouped.get_group(2000) # .get_group(param or group key) 
# takes 1 parameter the group name we want to see and returns the group data as a dataframe.


# print(yearly_movies_grouped.get_group(2000))

yearly_movies_grouped.groups.keys()
```

```python
group_keys = yearly_movies_grouped.groups.keys() # RETRIEVE KEYS FROM GROUPS
```

```python
with pd.ExcelWriter("netflix_by_years.xlsx") as writer:

    for key in group_keys:
#         print(key, "--> ", yearly_movies_grouped.get_group(key).shape)
        dataframe = yearly_movies_grouped.get_group(key)
        dataframe.to_excel(writer, sheet_name = f"{key}", index = False)
    
```
