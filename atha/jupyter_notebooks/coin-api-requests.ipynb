```python
## PRINTS SELECTED FIELDS FOR ALL DATA IN DATASET
for currency_dict in data:
#     print(currency_dict)
    print(currency_dict.get("name", "N/A"), currency_dict.get("data_start", "N/A"), currency_dict.get("price_usd", "N/A"))
    print("=========================================")
```

```python
## PRINTS ONLY FIRST 50 FIELDS FOR ALL DATA IN DATASET USING A COUNTER

i = 0

for currency_dict in data:

    print(currency_dict.get("name", "N/A"), currency_dict.get("data_start", "N/A"), currency_dict.get("price_usd", "N/A"))
    
    if i == 50:
        break
        
    i+=1
```

```python
# WRITE ALL DATA INTO A CSV FILE
file_name = "crypto.csv"
file = open(file_name, "w")
```

```python
i = 0

for currency_dict in data:

    file.write(f'{currency_dict.get("name", "N/A")}, {currency_dict.get("data_start", "N/A")}, {currency_dict.get("price_usd", "N/A")}\n')
    
    if i == 50:
        break
        
    i+=1
```

```python
file.close()
```
