```python
n = 0     # to get the wind

lenght = len(data.get("list"))   

while n<lenght:
    date = data.get("list")[n].get("dt_txt")
    temp = data.get("list")[n].get("main").get("temp") 
    wind = data.get("list")[n].get("wind").get("speed") 
    weather = data.get("list")[n].get("weather")[0].get("description")
    n+=1
# #     print(date, temp, wind)   #Adopt this if you wanna print things in columns
#     print(date.center(15),str(temp).center(15), str(wind).center(15), weather.rjust(15))
    print(date.center(15),str(temp).center(15), str(wind).center(15), weather.rjust(15))
    #so now we see how to separate using the .center, .rjust, ljust function.
    #.center by 15whitespaces, rjust = right justify
    #we can only use .center and .rjust functions on ONLY strings. 
    #Integers and floats have no attributes .center,that's why we converted all integer data to str
    
```

```python
file_name = "weather.csv"        #step 1: name the file
file = open(file_name, "w")      #step 2: open the file

n = 0                            #step 3: write the file in a loop.

lenght = len(data.get("list"))   

while n<lenght:
    date = data.get("list")[n].get("dt_txt")
    temp = data.get("list")[n].get("main").get("temp") 
    wind = data.get("list")[n].get("wind").get("speed") 
    weather = data.get("list")[n].get("weather")[0].get("description")
    n+=1
    file.write(f"{date.center(15)},{str(temp).center(15)}, {str(wind).center(15)}, {weather.rjust(15)} \n")
    
file.close()                   #step 4: close the file and run it. 
#Note the indentation of the file.close().
#N/b: the end of quote is after the \n. 
#The \n is necessary to run each data line by line. if not, it'll print in ONE long straight line

                               
```
